```python
import numpy as np
import pandas as pd
from pathlib import Path
import os.path
import csv
import cv2
from sklearn.model_selection import train_test_split

import tensorflow as tf
import pandas as pd
df = pd.read_csv ('/content/drive/MyDrive/Colab Notebooks/Amloki-classified/rating.csv',usecols= ['Name','Average'])
from sklearn.metrics import r2_score
path = Path('/content/drive/MyDrive/Colab Notebooks/Amloki-classified/Combined')
l=[]  
d=[]
for img in os.listdir(path):
        temp = os.path.splitext(img)[0]
        img_path = os.path.join(path, img)
        label=df.loc[df['Name']==temp]["Average"]
        t=list(label)
        arr = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
        new_arr = cv2.resize(arr, (224, 224))
        d.append([new_arr,t[0]])
print(d)        
```

```python
X1 = []
Y1 = []

IMG_SIZE=224
for features, label in d:
    X1.append(features)
    Y1.append(label)

X1 = np.array(X1).reshape(-1,IMG_SIZE,IMG_SIZE,1)
Y1 = np.array(Y1)
print(X1)
print(Y1)
```

```python
import pickle
pickle.dump(X1, open('/content/drive/MyDrive/X2.pkl', 'wb'))
pickle.dump(Y1, open('/content/drive/MyDrive/Y2.pkl', 'wb'))
```

```python
import pickle
pickle_in=open('/content/drive/MyDrive/X2.pkl', 'rb')
X=pickle.load(pickle_in)

pickle_in=open('/content/drive/MyDrive/Y2.pkl', 'rb')
y=pickle.load(pickle_in)
print(len(X))
print(len(y))
```

```python
from sklearn.model_selection import train_test_split
(trainX, testX, trainY, testY) = train_test_split(X, y,
	test_size=0.20, stratify=y, random_state=42)

print("Train Set Length: {} \nTest Set Length: {}".format(len(trainY), len(testY)))

print(trainY)
print(testY)
```

```python
from keras.preprocessing.image import ImageDataGenerator
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D
from keras.layers import Activation, Dropout, Flatten, Dense
from keras import backend as K
from tensorflow.python.keras import regularizers
train_data_dir = 'data/train'
validation_data_dir = 'data/validation'
nb_train_samples = 251
nb_validation_samples = 28
epochs = 50
batch_size = 10

trainx=trainX/255.0
```

```python
model = Sequential()
model.add(Conv2D(64, (3,3), input_shape=trainx.shape[1:]))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(3, 3)))

model.add(Conv2D(32, (3, 3), kernel_regularizer=regularizers.l2(l=0.10)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(64, (3, 3), kernel_regularizer=regularizers.l2(l=0.25)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Flatten())
model.add(Dense(256))
model.add(Activation('relu'))
model.add(Dropout(0.2))
model.add(Dense(1))
model.add(Activation('linear'))
model.compile(
    optimizer='adam',
    loss='mse'
)    

history=model.fit(trainX,trainY,batch_size,epochs,validation_split=0.15)
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import  roc_auc_score
import tensorflow as tf
import numpy as np
from tensorflow import keras

y_pred=model.predict(testX)
y_predict=np.ceil(y_pred)
y_predict=y_predict.astype(int)
print(y_predict)
print(testY)
```

```python
from sklearn.metrics import mean_squared_error

rms = mean_squared_error(testY, y_predict,squared=False)
print(rms)
```
